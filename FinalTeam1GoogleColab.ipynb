# **5. Import neccesary libaries**

These libraries work together to enable key tasks such as PDF text extraction, AI-driven dialogue generation, and seamless interaction with language models. Installing and importing them at the beginning ensures the system is ready to process user input and generate meaningful conversations.

```python
import os
import time
from langchain.chat_models import init_chat_model
from langchain.schema import SystemMessage, HumanMessage, AIMessage
```

# **6. Enter GROQ API key**

To access and utilize the language model services provided by Groq, users must supply a valid API key. The system prompts the user to enter this key, which is then stored securely in an environment variable for use during runtime. This authentication step is required to initialize the AI agents and generate conversational responses based on the uploaded document.

```python
apik = input("Please enter your GROQ API key: ").strip()
if not apik:
    raise ValueError("Error: API key cannot be empty.")
else:
    os.environ['GROQ_API_KEY'] = apik
```

# **7. Using two llama3 models for alpha and beta**

To create a rich, dynamic, and insightful academic conversation, the CSUSB Study Podcast Assistant uses two separate AI agents‚ÄîAlpha and Beta‚Äîeach powered by distinct variants of the LLaMA 3 language model. This dual-model setup enhances the diversity of thought and tone in the generated dialogue.

```python
chat_alpha = init_chat_model("llama3-8b-8192", model_provider="groq")  # Alpha's model
chat_beta = init_chat_model("llama3-70b-8192", model_provider="groq")  # Beta's model

messages = [SystemMessage(content="You are an AI assistant that will help.")]
```

# **8. Asking question to chatbot**

The CSUSB Study Podcast Assistant allows users to directly engage with the AI agents by submitting custom questions related to the uploaded academic content. This feature enhances interactivity and provides tailored support for studying or clarification.

```python
user_input = input("Ask the Chatbot a Question: ")

if user_input:
    prompt = f"User Query: {user_input}"

    response = chat_alpha.invoke([SystemMessage(content=prompt)])
    ai_response = response.content if response else "I do not know!"

    print(f"Alpha: {user_input}")
    print(f"Beta: {ai_response}")
```

# **9. Conversation between alpha and beta regarding csusb (aipodcast)**

The **start_aipodcast()** function simulates a short, dynamic podcast-style dialogue between two AI agents, Alpha and Beta, centered around California State University, San Bernardino (CSUSB). This automated session highlights how the system can generate meaningful, topic-driven conversations using LLaMA 3 language models

```python
def start_aipodcast():
    for _ in range(10):  # 10 questions session
        # Alpha asks a question about CSUSB
        question_prompt = "Generate a short, clear, and direct question (1-3 lines) about CSUSB's podcast, events, admissions, or campus life."

        messages.append(HumanMessage(content=question_prompt))
        question_response = chat_alpha.invoke(messages)
        question = question_response.content.strip() if question_response else "Could not generate a question."

        print(f"Alpha: {question}")
        time.sleep(1)

#Beta answers in 3-4 lines summary format
        beta_prompt = f"Provide a concise 3-4 line summary answering the following question: {question}"
        response = chat_beta.invoke([SystemMessage(content=beta_prompt)])
        ai_response = response.content.strip() if response else "I don't know"

        messages.append(AIMessage(content=ai_response))
        print(f"Beta: {ai_response}")
        print("-" * 50)
        time.sleep(1)

#Start AI Podcast
start_aipodcast()
```
